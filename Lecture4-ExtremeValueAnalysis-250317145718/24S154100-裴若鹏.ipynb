Using the parameter estimation approach in code GB 50009 2012, predict the 10, 20, 50, 100, 200, 500, 1000 years return period design wind speeds for the above two cases, and compare the results with the Gumbel approach.

# Answer


```python
import numpy as np
sorted_df=df.sort_values(by="Wind speed m/s")
new_data1=sorted_df
new_data1=new_data1.reset_index(drop=True)
```

```python
hangshu=new_data1.shape[0]
p=[]
for i in range(1, hangshu+1):  
    p.append(i/(hangshu+1))  
new_data1["P"]=p
```

```python
new_data1["Y"]=-np.log(-np.log(new_data1["P"]))
# 使用 NumPy 的 polyfit 方法拟合直线
# deg=1 表示拟合一次多项式（直线）
coefficients = np.polyfit(new_data1["Y"], new_data1["Wind speed m/s"], deg=1)
# 获取拟合的斜率和截距
k1= coefficients[0]
b1= coefficients[1]
```

```python
R=[10,20,50,100,200,500,1000]#回归周期
U=0
ExtremeValue1=[]
for i in R:
    U=b1-k1*np.log(-np.log(1-1/i))
    ExtremeValue1.append(U)
ExtremeValue1=np.around(ExtremeValue1,decimals=2)
```

```python
new_data2=new_data1
new_data2= new_data2.drop(new_data2.index[-1])   
```

```python
coefficients = np.polyfit(new_data2["Y"], new_data2["Wind speed m/s"], deg=1)
# 获取拟合的斜率和截距
k2 = coefficients[0]
b2= coefficients[1]
R=[10,20,50,100,200,500,1000]#回归周期
U=0
ExtremeValue2=[]
for i in R:
    U=b2-k2*np.log(-np.log(1-1/i))
    ExtremeValue2.append(U)
ExtremeValue2=np.around(ExtremeValue2,decimals=2)
```

```python
new_data3=new_data1
hangshu=new_data3.shape[0]
p=[]
for i in range(1, hangshu+1):  
    p.append((i-0.44)/(hangshu+0.12))  
new_data3["P"]=p
new_data3["Y"]=-np.log(-np.log(new_data3["P"]))
coefficients = np.polyfit(new_data3["Y"], new_data3["Wind speed m/s"], deg=1)
# 获取拟合的斜率和截距
k3= coefficients[0]
b3= coefficients[1]
R=[10,20,50,100,200,500,1000]#回归周期
U=0
ExtremeValue3=[]
for i in R:
    U=b3-k3*np.log(-np.log(1-1/i))
    ExtremeValue3.append(U)
ExtremeValue3=np.around(ExtremeValue3,decimals=2)
```

```python
new_data4=new_data2
hangshu=new_data4.shape[0]
p=[]
for i in range(1, hangshu+1):  
    p.append((i-0.44)/(hangshu+0.12))  
new_data4["P"]=p
new_data4["Y"]=-np.log(-np.log(new_data4["P"]))
coefficients = np.polyfit(new_data4["Y"], new_data4["Wind speed m/s"], deg=1)
# 获取拟合的斜率和截距
k4= coefficients[0]
b4= coefficients[1]
R=[10,20,50,100,200,500,1000]#回归周期
U=0
ExtremeValue4=[]
for i in R:
    U=b4-k4*np.log(-np.log(1-1/i))
    ExtremeValue4.append(U)
ExtremeValue4=np.around(ExtremeValue4,decimals=2)
```

```python
ExtremeValue1= pd.DataFrame(ExtremeValue1, columns=['wind speed m/s'])
ExtremeValue1.insert(0, 'Return period', [10, 20, 50,100,200,500,1000]) 
ExtremeValue2= pd.DataFrame(ExtremeValue2, columns=['wind speed m/s'])
ExtremeValue2.insert(0, 'Return period', [10, 20, 50,100,200,500,1000]) 
ExtremeValue3= pd.DataFrame(ExtremeValue3, columns=['wind speed m/s'])
ExtremeValue3.insert(0, 'Return period', [10, 20, 50,100,200,500,1000])
ExtremeValue4= pd.DataFrame(ExtremeValue4, columns=['wind speed m/s'])
ExtremeValue4.insert(0, 'Return period', [10, 20, 50,100,200,500,1000]) 
print(f"用Gumbel法在情况1下拟合的直线方程为：y = {k1:.2f}x + {b1:.2f}")
print("用Gumbel法在情况1下预测的最大风速为:")
print(f"{ExtremeValue1}")
print(f"用Gumbel法在情况2下拟合的直线方程为：y = {k2:.2f}x + {b2:.2f}")
print("用Gumbel法在情况2下预测的最大风速为:")
print(f"{ExtremeValue2}")
print(f"用Gringorten法在情况1下拟合的直线方程为：y = {k3:.2f}x + {b3:.2f}")
print("用Gringorten法在情况1下预测的最大风速为：")
print(f"{ExtremeValue3}")
print(f"用Gringorten法在情况2下拟合的直线方程为：y = {k4:.2f}x + {b4:.2f}")
print("用Gringorten法在情况2下预测的最大风速为：")
print(f"{ExtremeValue4}")
text1="观察以上数据可以发现，在四组数据中，1000年回归周期的风速预测极值均大于50年回归周期的风速预测极值。\
\n两种方法中，去除掉最大值后，风速预测值都减小，说明某一年的极大值对整体的风速预测值有较大影响\
\n两种情况下，相同回归周期，Gringorten法预测的风速极值小于Gumbel法预测的风速极值，说明Gringorten法对P的计算作了修正后使得风速预测极值更为合理"
print(text1)
```

```python
n=[10,20,50,100,200,500,1000]
C1=[0.9497,1.06283,1.16066,1.20649,1.24292,1.2588,1.26851]
C2=[0.4952,0.52355,0.54853,0.56002,0.56878,0.57240,0.57450]
xishu=[]
xishu=pd.DataFrame(xishu)
xishu["return period"]=n
xishu["C1"]=C1
xishu["C2"]=C2
```

```python
# 计算列'风速'的均值
junzhi= new_data1['Wind speed m/s'].mean()
# 计算列'A'的标准差
biaozhuncha= new_data1['Wind speed m/s'].std()
xishu["α"]=xishu["C1"]/biaozhuncha
xishu["u"]=junzhi-xishu["C2"]/xishu["α"]
xishu["extreme wind speed"]=xishu["u"]-1/xishu["α"]*np.log(-np.log(1-1/xishu["return period"]))

xishu2=[]
xishu2=pd.DataFrame(xishu2)
xishu2["return period"]=n
xishu2["C1"]=C1
xishu2["C2"]=C2
# 计算列'风速'的均值
junzhi2= new_data2['Wind speed m/s'].mean()
# 计算列'A'的标准差
biaozhuncha2= new_data2['Wind speed m/s'].std()
xishu2["α"]=xishu2["C1"]/biaozhuncha2
xishu2["u"]=junzhi2-xishu2["C2"]/xishu2["α"]
xishu2["extreme wind speed"]=xishu2["u"]-1/xishu2["α"]*np.log(-np.log(1-1/xishu2["return period"]))
print("国标方法情况1下计算出的极值风速：")
print(xishu)
print("国标方法情况2下计算出的极值风速：")
print(xishu2)
text2="国标法与Gumbel法相比，当不忽略最大风速值时，两种方法预测的最大风速值差别较小，可以认为基本相同。\
\n当忽略风速极大值时，10、20、50年风速预测值差别较小，100、200、500、1000年风速预测差值在1~2m/s,且Gumbel法得出的风速值较大"
print(text2)
```
