```python
# 导入必要的库，pandas用于数据处理
import pandas as pd
from io import StringIO # python3中StringIO的位置
import numpy as np
import matplotlib.pyplot as plt

# 加载数据
data = """
year,Wind speed m/s
1952,31.4
1953,33.4
1954,29.8
1955,30.3
1956,27.8
1957,30.3
1958,29.3
1959,36.5
1960,29.3
1961,27.3
1962,31.9
1963,28.8
1964,25.2
1965,27.3
1966,23.7
1967,27.8
1968,32.4
1969,27.8
1970,26.2
1971,30.9
1972,31.9
1973,27.3
1974,25.7
1975,32.9
1976,28.3
1977,27.3
1978,28.3
1979,28.3
1980,29.3
1981,27.8
1982,27.8
1983,30.9
1984,26.7
1985,30.3
1986,28.3
1987,30.3
1988,34
1989,28.8
1990,30.3
1991,27.3
1992,27.8
1993,28.8
1994,30.9
1995,26.2
1996,25.7
1997,24.7
1998,42.2
"""

# 将data数据转换为StringIO对象，再读取为DataFrame
df = pd.read_csv(StringIO(data))

# 提取最大风速
max_wind_speeds = df['Wind speed m/s']

# 计算含1998年数据的风速均值和标准差
mean_with_1998 = max_wind_speeds.mean()
std_with_1998 = max_wind_speeds.std()

# 排除最后一年的风速
max_wind_speeds_without_1998 = max_wind_speeds[:-1]

# 计算不含1998年数据的风速均值和标准差
mean_without_1998 = max_wind_speeds_without_1998.mean()
std_without_1998 = max_wind_speeds_without_1998.std()

# 输出结果
print(f"含1998年数据的风速均值: {mean_with_1998:.2f}, 标准差: {std_with_1998:.2f}")
print(f"不含1998年数据的风速均值: {mean_without_1998:.2f}, 标准差: {std_without_1998:.2f}")

```
