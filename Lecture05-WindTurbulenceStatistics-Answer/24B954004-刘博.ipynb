### 1 Statistical results of the z-direction wind velocity time history

```python
import scipy.io as sio
import numpy as np  
import matplotlib.pyplot as plt  
from scipy import signal  
from scipy.stats import norm  

# load data
simDataPath = './windData/zDirData.mat'
simData = sio.loadmat(simDataPath, squeeze_me=True, struct_as_record=False)
U = simData['U']
V = simData['V']
W = simData['W']
X = simData['X']
Y = simData['Y']
Z = simData['Z']
dt = simData['dt']
del simData
```

#### 1.1 Mean velocity profile

```python
Y_positions = [0, 20, 40, 60, 80]  
heights_z = [10, 30, 50, 70, 90]  

def plot_mean_velocity(U, heights, Ur=50, zr=160, alpha=0.22):
    U_avg_measured = np.mean(U, axis=1)
    z_values = np.array(heights)
    U_avg_theory = Ur * (z_values / zr)**alpha

    plt.figure(figsize=(8, 6))
    plt.scatter(U_avg_measured, heights, c='blue', label='Measured', s=100, edgecolor='black')
    plt.plot(U_avg_theory, heights, 'r-.', label='Theory', linewidth=2.5)
    plt.xlabel('Mean Velocity (m/s)', fontsize=12, fontweight='bold')
    plt.ylabel('Height (m)', fontsize=12, fontweight='bold')
    plt.title('Mean Velocity Profile (z-direction)', fontsize=14, fontweight='bold')
    plt.legend(fontsize=12, frameon=True, shadow=True)
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.gca().invert_yaxis()
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

plot_mean_velocity(U, heights_z)
```

#### 1.2 Turbulence intensity profile

```python
def turbulence_intensity(U, V, W, heights, I10=0.23, alpha=0.22):
    u_prime = U - np.mean(U, axis=1, keepdims=True)
    v_prime = V - np.mean(V, axis=1, keepdims=True)
    w_prime = W - np.mean(W, axis=1, keepdims=True)

    Iu_measured = np.std(u_prime, axis=1) / np.mean(U, axis=1)
    Iv_measured = np.std(v_prime, axis=1) / np.mean(U, axis=1)
    Iw_measured = np.std(w_prime, axis=1) / np.mean(U, axis=1)

    z = np.array(heights)
    Iu_theory = I10 * (z / 10)**(-alpha)
    Iv_theory = 0.78 * Iu_theory
    Iw_theory = 0.55 * Iu_theory

    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    for ax, I_measured, I_theory, title, color in zip(axs, 
                                                     [Iu_measured, Iv_measured, Iw_measured], 
                                                     [Iu_theory, Iv_theory, Iw_theory], 
                                                     ['Iu(z)', 'Iv(z)', 'Iw(z)'], 
                                                     ['blue', 'green', 'red']):
        ax.scatter(I_measured, heights, c=color, label='Measured', s=100, edgecolor='black', alpha=0.8)  # 散点图
        ax.plot(I_theory, heights, 'm-.', label='Theory', linewidth=2.5)  
        ax.set_title(title, fontsize=14, fontweight='bold')
        ax.set_xlabel('Turbulence Intensity', fontsize=12, fontweight='bold')
        ax.set_ylabel('Height (m)', fontsize=12, fontweight='bold')
        ax.grid(True, linestyle=':', alpha=0.7)
        ax.legend(fontsize=12, frameon=True, shadow=True)
        ax.tick_params(axis='both', labelsize=10)

    plt.tight_layout()
    plt.show()

turbulence_intensity(U, V, W, heights_z)
```

#### 1.3 Wind power spectral density

```python
def wind_spectrum(u_prime, dt, L_u, U_avg, component='u'):
    fs = 1 / dt
    f, Su = signal.welch(u_prime, fs, nperseg=1024)

    if component == 'u':
        Su_theory = (4 * (L_u * U_avg)**2 * (L_u / U_avg)) / \
                   (1 + 70.8 * (L_u * f / U_avg)**2)**(5/6)
    elif component in ['v', 'w']:
        Su_theory = (4 * (L_u * U_avg)**2 * (L_u / U_avg) * (1 + 188.4 * (2 * L_u * f / U_avg)**2)) / \
                   (1 + 70.8 * (2 * L_u * f / U_avg)**2)**(11/6)

    plt.figure(figsize=(8, 6))
    plt.loglog(f, Su, 'b', label='Measured', linewidth=2.5)
    plt.loglog(f, Su_theory, 'r-.', label='Theory', linewidth=2.5)
    plt.xlabel('Frequency (Hz)', fontsize=12, fontweight='bold')
    plt.ylabel('Power Spectral Density', fontsize=12, fontweight='bold')
    plt.title(f'{component.upper()}-component Spectrum at z=50m', fontsize=14, fontweight='bold')
    plt.legend(fontsize=12, frameon=True, shadow=True)
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.tick_params(axis='both', labelsize=10)
    plt.tight_layout()
    plt.show()

idx = 2
u_prime = U[idx] - np.mean(U[idx])
L_u = 100 * (heights_z[idx] / 30)**0.5  
wind_spectrum(u_prime, dt=0.01, L_u=L_u, U_avg=np.mean(U[idx]), component='u')
```

#### 1.4 Probability density curve

```python
def plot_pdf(u_prime):
    hist, bins = np.histogram(u_prime, bins=50, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    mu, std = norm.fit(u_prime)
    pdf_theory = norm.pdf(bin_centers, mu, std)

    plt.figure(figsize=(8, 6))
    plt.plot(bin_centers, hist, 'b-', label='Measured', linewidth=2)
    plt.plot(bin_centers, pdf_theory, 'r--', label='Normal Fit', linewidth=2)
    plt.xlabel('Fluctuating Velocity (m/s)', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.title('PDF of u-component at z=50m', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tick_params(axis='both', labelsize=10)
    plt.tight_layout()
    plt.show()

plot_pdf(u_prime)
```

### 2 Statistical results of the y-direction wind velocity time history

```python
# load data
simDataPath = './windData/yDirData.mat'
simData = sio.loadmat(simDataPath, squeeze_me=True, struct_as_record=False)
U = simData['U']
V = simData['V']
W = simData['W']
X = simData['X']
Y = simData['Y']
Z = simData['Z']
dt = simData['dt']
del simData
```

#### 2.1 Spatial correlation coefficient in y direction

```python
def spatial_correlation(U, Y_positions, L_prime=60):
    ref_u = U[0]
    correlations = []
    distances = []

    for i in range(1, 5):
        target_u = U[i]
        corr = np.corrcoef(ref_u, target_u)[0, 1]
        correlations.append(corr)
        distances.append(Y_positions[i] - Y_positions[0])

    r = np.array(distances)
    rho_theory = np.exp(-r / L_prime)

    plt.figure(figsize=(8, 6))
    plt.plot(distances, correlations, 'bo-', label='Measured', markersize=10, linewidth=2.5)
    plt.plot(distances, rho_theory, 'r-.', label='Theory', linewidth=2.5)
    plt.xlabel('Transverse Distance (m)', fontsize=12, fontweight='bold')
    plt.ylabel('Correlation Coefficient', fontsize=12, fontweight='bold')
    plt.title('Spatial Correlation in y-direction (u-component)', fontsize=14, fontweight='bold')
    plt.legend(fontsize=12, frameon=True, shadow=True)
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.tick_params(axis='both', labelsize=10)
    plt.tight_layout()
    plt.show()

spatial_correlation(U, Y_positions, L_prime=60)
```
